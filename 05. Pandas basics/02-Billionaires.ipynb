## 4a) How many male billionaires are there compared to the number of female billionares?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
df['gender'].value_counts()
```

## 4b) What percent is that?

```python
males = df['gender'].value_counts()['male']
females = df['gender'].value_counts()['female']

print(f"The {round(males / (males+females) * 100)} percent of the billionares are male.")
```

```python
df['gender'].value_counts(normalize=True) * 100
```

## 4c) Is the average wealth different for a female billionaire as compared to a male billionaire?

```python
df.groupby(['gender'])['networthusbillion'].mean()
```

```python
male_av = df.loc[df['gender'] == 'male', 'networthusbillion'].mean()
female_av = df.loc[df['gender'] == 'female', 'networthusbillion'].mean()

print(f"The average wealth for female billionare is {(female_av):.3}, while for men billionaires is {(male_av):.3}, which are pretty similar!")
```
