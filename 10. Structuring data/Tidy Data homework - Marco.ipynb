```python
import pandas as pd
```

```python
df = pd.read_csv('Data Tidying Examples - Sidewinder Library.csv')
df
```

### First, I will create ID numbers for members and books

This will allow me to do calculations without worrying of duplicate values.

```python
df['member_id'] = (
    df['First Name']
    .str[0:3]+df['Last Name']
    .str[0:3]+df['Date of Birth']
    .str[0:4]
).str.upper()

df['book_id'] = (
    df['Author_A']
    .str[0:5]+df['Book']
    .str[0:5]
).str.upper()

df
```

### Then, I will create new datasets with specific informations that belongs to our observational units

#### The first observational unit is the single book and its characteristics

```python
books = df [[
    'book_id',
    'Book',
    'Author_A',
    'Author_B',
    'Genres'
]]

books
```

#### Because the column `Genres` include more values, I will need to explode it

```python
books['Genres'] = books['Genres'].str.split(pat=';')
books = books.explode('Genres', ignore_index=True)
books
```

#### Because the columns `Author_A` and `Author_B` both include values of the same variable, I will need to combine them and then explode them.

```python
books['author'] = books['Author_A']+';'+books['Author_B'].str.replace('-',"")
books['author'] = books['author'].str.split(pat=';')
books = books.explode('author', ignore_index=True)
books = books.drop('Author_A', axis=1)
books = books.drop('Author_B', axis=1)

books
```

#### I will now get rid of rows that do not contain an Author. Because those are empty strings and not `null` values, `dropna()` won't work.

```python
books = books[books['author'] != '']
books
```

```python
books = books[books['author'] != '']
books
```

```python
#Let's rename our columns

books = books.rename(columns={"Book": "book", "Genres": "genre"})
books
```

```python
# And save it to a .csv file
books.to_csv('books.csv')
```

### Second observational unit: members of the library
This should be faster as I only have four or five elements to take into consideration.

```python
members = df [[
    'member_id',
    'First Name',
    'Last Name',
    'Date of Birth',
    'Library Card Expired?'
]]

members = members.drop_duplicates()
members
```

```python
# Rename columns
members = members.rename(columns={
    "First Name": "first_name",
    "Last Name": "last_name",
    "Date of Birth" : "date_of_birth",
    "Library Card Expired?" : "card_expired"
})
members
```

```python
# Save to .csv file
members.to_csv('members.csv')
```
